```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
```

```python
from selenium.webdriver.common.by import By
```

```python
driver = webdriver.Chrome()
#크롬 드라이버에 url 주소 넣고 실행
driver.get("https://google.co.kr")
# time.sleep(3)
search_box = driver.find_element(By.XPATH, '//*[@id="APjFqb"]');
search_box.send_keys("**Hello**")
while True:
    time.sleep(1)
```

```python
df=pd.DataFrame({"a":[1,2,3,4],"b":[11,22,22,33],"c":[6,6,7,7]})
df
```

```python
df.groupby('c').sum() # 6: 11+22, 7: 22+33
```

```python
def f(i):
    return "lucky" + str(sum(i)+3)
```

```python
def g(i):
    return str(i.mean()) + "G"
```

```python
df.groupby('c').agg(['sum', f, g]) # aggregation
```
